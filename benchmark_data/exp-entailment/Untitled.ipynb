```python
import pandas as pd
import random

df = pd.read_csv('output2.csv')
```

```python
df = df[df['label']==1]
```

```python
df = df.iloc[::5]
```

```python
df = df.reset_index(drop=True)
df
```

```python
# Function to generate negative samples
def generate_negative_samples(df):
    negative_samples = []
    for index, row in df.iterrows():
        # Randomly select a different 'sent2' to create a negative sample
        random_index = random.choice([i for i in range(len(df)) if i != index])
        negative_row = row.copy()
        negative_row['sent2'] = df.loc[random_index, 'sent2']
        negative_row['label'] = 0
        negative_samples.append(negative_row)
    
    # Convert list of negative samples to a DataFrame and append it
    negative_samples_df = pd.DataFrame(negative_samples)
    df = pd.concat([df, negative_samples_df], ignore_index=True)
    
    return df

# Generate negative samples and append them to the dataframe
df_with_negatives = generate_negative_samples(df)
df_with_negatives
```

```python
df_with_negatives.to_csv('detailed_description_only.csv', index=False)
```
